```python
def get_title(name):
    title = ""
    name_list = name.split()
    for thing in name_list:
        if '.' in thing:
            title = thing

    title_dict = {
        'Mr.': 'Mr',
        'Mrs.': 'Mrs',
        'Miss.': 'Miss',
        'Master.': 'Master',
        'Don.': 'Noble',
        'Rev.': 'Noble',
        'Dr.': 'Noble',
        'Mme.': 'Mrs',
        'Ms.': 'Mrs',
        'Major.': 'Military',
        'L.': 'Military',
        'Lady.': 'Noble',
        'Sir.': 'Noble',
        'Mlle.': 'Miss',
        'Col.': 'Military',
        'Capt.': 'Military',
        'Countess.': 'Noble',
        'Jonkheer.': 'Noble',
        'Dona.': 'Noble'
        }

    return title_dict[title]
```

```python
train_dataframe = raw_train_dataframe
test_dataframe = raw_test_dataframe
all_data = pd.concat([train_dataframe, test_dataframe])

for df in [train_dataframe, test_dataframe]:
    df['Title'] = df.Name.map(get_title)
    df['GroupSize'] = df['Ticket'].map(all_data['Ticket'].value_counts())
    df.Fare = df.Fare.fillna(30)
    df['FarePerPerson'] = df['Fare']/df['GroupSize']
    df.Age = df.Age.fillna(df.groupby(['Pclass', 'Parch', 'SibSp', 'Title']).Age.transform('median'))
    df.Embarked = df.Embarked.fillna('C')
    #df.Age = df.Age.fillna(df.Age.median())
    df['Age*Class'] = df['Age']*df['Pclass']

```

```python
k_fold = StratifiedKFold(n_splits=5)
gridcv = GridSearchCV(model, param_grid=params, cv=k_fold, scoring='accuracy', n_jobs=4, verbose=1)

gridcv.fit(X_train, y_train)

best_model = gridcv.best_estimator_
```

```python
model1 = best_model

model1.fit(X_train, y_train)
y_predict = model1.predict(X_test)
output = pd.DataFrame({'PassengerId': test_dataframe.PassengerId, 'Survived': y_predict})
output.to_csv('submission2.csv', index=False)
```
